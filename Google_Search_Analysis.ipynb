# Google Search Analysis with Python

Google Trends provides an API that can be used to analyze the daily searches on Google and it also knows as pytrends. You can simply download it using pip command; pip install _pytrends_.

```python
# !pip install pytrends
```

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

Here I will be analysing the Google search trends on the queries based on 'Starbucks', so let's create a DataFrame of the top 10 countries searching for 'Starbucks' on Google:

```python
trends.build_payload(kw_list=['Starbucks'])
data = trends.interest_by_region()
data = data.sort_values(by='Starbucks', ascending=False)
data = data.head(10)
data
```

```python
data.reset_index().plot(x='geoName', y='Starbucks', figsize=(9, 6), kind='bar')

plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Starbucks'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(9, 6))
data['Starbucks'].plot()
plt.title('Total Google Searches for Starbucks', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

We can see that there is a common high trend at the end of the years and it looks like it tends to decrease.
